```python
import numpy as np
import cv2
import keras
from keras.models import *
from keras.layers import *
from keras import callbacks
from keras.applications import MobileNetV2
from keras.preprocessing.image import ImageDataGenerator
import shutil
import os
```

```python
import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
keras.backend.tensorflow_backend.set_session(tf.Session(config=config))
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
train_data_dir = '/Users/yh-mac/Downloads/Bald_Classification/data/Dataset/Train/'
valivation_data_dir = '/Users/yh-mac/Downloads/Bald_Classification/data/Dataset/Validation/'

img_height = 64
img_width = 64
class_num = 2
batch_size = 64
epochs = 20

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(
    rescale=1./255
)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_height, img_width),
    batch_size=32
)

validation_generator = validation_datagen.flow_from_directory(
    valivation_data_dir,
    target_size=(img_height, img_width),
    batch_size=32
)

num_train = train_generator.n
num_validation = validation_generator.n

steps_per_epoch = num_train // batch_size
steps_per_epoch_val = num_validation // batch_size
```

```python
steps_per_epoch, steps_per_epoch_val
```

```python
net = MobileNetV2(
    input_shape=(img_height, img_width, 3),
    include_top=False,
    alpha=0.3,
    weights=None
)
x = net.output
x = GlobalAveragePooling2D()(x)
x = Dense(class_num)(x)
out = Activation("softmax")(x)
model = Model(inputs = [net.input] , outputs = [out] )
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=["accuracy"])
cbks = [callbacks.ModelCheckpoint("best_bald_classifity.h5", save_best_only=True),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, verbose=1, mode='auto', epsilon=0.001, cooldown=0, min_lr=1e-6)]
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    validation_data=validation_generator,
    nb_val_samples=steps_per_epoch_val,
    callbacks=cbks
)
```

```python
model.save('bald_classifity.h5')
```
